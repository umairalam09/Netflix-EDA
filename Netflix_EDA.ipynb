```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/netflix_titles.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['director','cast'],inplace = True, axis = 1)
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.describe()
```

```python
df.fillna({'director': 'unknown', 'cast': 'unknown', 'country':'unknown'}, inplace = True)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x = 'type', data = df, palette= 'coolwarm')
plt.title("Distribution of Content Type (Movies vs TV Shows)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(x = 'release_year', data = df, palette= 'coolwarm', bins = 30, kde = True)
plt.title("Distribution of Ratings")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(y=df['rating'], order=df['rating'].value_counts().index, palette='magma')
plt.title("Distribution of Content Ratings")
plt.show()
```

```python
df['listed_in'] = df['listed_in'].astype(str).fillna('')

# Split genres properly
all_genres = [genre for sublist in df['listed_in'].dropna().str.split(', ') for genre in sublist]

# Count occurrences
genre_counts = pd.Series(all_genres).value_counts()


plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.values[:10], y=genre_counts.index[:10], palette='viridis')
plt.title("Top 10 Most Popular Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

```python
df_yearly = df.groupby('release_year').size()
plt.figure(figsize=(12, 6))
sns.histplot(x=df_yearly.index, y=df_yearly.values)
plt.title("Number of Releases Per Year")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
country_counts = df['country'].value_counts().head(10)
sns.barplot(y=country_counts.index, x=country_counts.values, palette='pastel')
plt.title("Top 10 Countries Producing Netflix Content")
plt.show()

```

```python
df['duration'] = df['duration'].str.extract('(\d+)').astype(float)  # Extract numeric values
plt.figure(figsize=(10, 5))
sns.histplot(df['duration'].dropna(), bins=20, kde=True)
plt.title("Distribution of Content Duration (Movies & TV Shows)")
plt.show()

```

```python
numerical_features = ['release_year', 'duration']
sns.heatmap(df[numerical_features].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
print("\nKey Findings:")
print("1. Most content on Netflix is Movies rather than TV Shows.")
print("2. Netflix has a diverse genre catalog, with Drama and Comedy being the most popular.")
print("3. Content releases have increased over time, peaking in recent years.")
print("4. USA dominates content production, followed by India and the UK.")
print("5. Most content falls within a reasonable duration range, but some variations exist.")
```
